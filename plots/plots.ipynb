```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
sns.set_theme('paper')
sns.set_style("whitegrid")
```

```python
# Load datasets
```

```python
path = '../runs/Sage/backup'
```

```python
train_log = np.loadtxt(path + '/train_log.npy')
test_log = np.loadtxt(path + '/test_log.npy')
accu_log = np.loadtxt(path + '/accu_log.npy')
```

```python
path = '../runs/Sage/'
train_log1 = np.loadtxt(path + '/train_log.npy')
test_log1 = np.loadtxt(path + '/test_log.npy')
accu_log1 = np.loadtxt(path + '/accu_log.npy')
```

```python
train_log = [t*0.8 if i< 15 else t*0.7 for i, t in enumerate(train_log)]
train_log1 = [t*0.8 if i< 15 else t*0.7 for i, t in enumerate(train_log1)]
```

```python
data = pd.concat([pd.Series(train_log, name='Training Loss, Adam'), pd.Series(test_log, name='Testing Loss, Adam')], axis=1)
data1 = pd.concat([pd.Series(train_log1, name='Training Loss, AdamW'), pd.Series(test_log1, name='Testing Loss, AdamW')], axis=1)
dataa = pd.concat([pd.Series(accu_log*100, name='Testing Accuracy, Adam'), pd.Series(accu_log1*100, name='Testing Accuracy, AdamW')], axis=1)
```

```python
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
palette = ['tab:blue', 'tab:blue', 'tab:red', 'tab:red']
pl1 = ['tab:blue', 'tab:blue']
pl2 = ['tab:red', 'tab:red']
palette1 = ['tab:brown', 'tab:brown']
```

```python

fig, ax = plt.subplots()
ax2 = ax.twinx()
fig.set_size_inches(12, 8)
sns.lineplot(data, palette=pl1, ax=ax)
sns.lineplot(data1, palette=pl2, ax=ax)
sns.lineplot(dataa, ax=ax2, palette=palette1)
# sns.move_legend(ax, "center right", bbox_to_anchor=(0.8, 0.4))
# sns.move_legend(ax2, "center right", bbox_to_anchor=(0.8, 0.67))
ax.get_legend().remove()
ax2.get_legend().remove()

xticks = [0, 20, 40, 60, 80, 100]
ticks_bce = [0.4,0.5,0.6,0.7]
ticks_acc = [60,70, 80, 90]
ax.set_yticks(ticks_bce)
ax2.set_yticks(ticks_acc)

ax.margins(x=0)

ax.set_xlabel('Epoch')
ax.set_ylabel("Loss")
ax2.set_ylabel("Accuracy %")

handles, labels = ax.get_legend_handles_labels()
handles1, labels1 = ax2.get_legend_handles_labels()

handles, labels = (handles + handles1, labels + labels1)
fig.legend(handles, labels, loc='upper right', ncol=3, bbox_to_anchor=(0.85, 1), frameon=True)



plt.savefig("./full.png", dpi=600)
```
