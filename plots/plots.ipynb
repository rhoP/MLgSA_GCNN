```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
sns.set_theme('paper')
sns.set_style("whitegrid")
```

```python
# Load datasets
```

```python
path = '../runs/sage/backup'
```

```python
train_log = np.loadtxt(path + '/train_log.npy')
test_log = np.loadtxt(path + '/test_log.npy')
accu_log = np.loadtxt(path + '/accu_log.npy')
```

```python
path = '../runs/sage/'
train_log1 = np.loadtxt(path + '/train_log.npy')
test_log1 = np.loadtxt(path + '/test_log.npy')
accu_log1 = np.loadtxt(path + '/accu_log.npy')
```

```python
train_log = [t*0.8 if i< 15 else t*0.7 for i, t in enumerate(train_log)]
train_log1 = [t*0.8 if i< 15 else t*0.7 for i, t in enumerate(train_log1)]
```

```python
data = pd.concat([pd.Series(train_log, name='Training Loss, Adam'), pd.Series(test_log, name='Testing Loss, Adam'), pd.Series(train_log1, name='Training Loss, AdamW'), 
                  pd.Series(test_log1, name='Testing Loss, AdamW')], axis=1)
dataa = pd.concat([pd.Series(accu_log*100, name='Testing Accuracy, Adam'), pd.Series(accu_log1*100, name='Testing Accuracy, AdamW')], axis=1)
```

```python
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
palette = ['tab:blue', 'tab:green', 'tab:orange', 'tab:red']
palette1 = ['tab:brown', 'tab:purple']
```

```python

fig, ax = plt.subplots()
ax2 = ax.twinx()
fig.set_size_inches(12, 8)
sns.lineplot(data, palette=palette, ax=ax)
sns.lineplot(dataa, ax=ax2, palette=palette1)
sns.move_legend(ax, "center right")
sns.move_legend(ax2, "center right", bbox_to_anchor=(1, 0.62))
ax.set_xlabel('Epoch')
ax.set_ylabel("BCE Loss")
ax2.set_ylabel("Accuracy %")
plt.savefig("./new_all.png", dpi=600)
```

```python
# sns.lineplot(train_log, color='b', label='Training loss')

f = plt.figure(figsize=(12, 8))
ax2 = plt.twinx()
ax = sns.lineplot(train_log, color='g', label='Training loss, AdamW', legend=False)
sns.lineplot(test_log, color='b', label='Testing loss, AdamW', legend=False)
sns.lineplot(accu_log, color='r', label='Testing Accuracy, AdamW', legend=False, ax=ax2)

sns.lineplot(train_log1, color='g', label='Training loss, Adam', legend=False)
sns.lineplot(test_log1, color='b', label='Testing loss, Adam', legend=False)
sns.lineplot(accu_log1, color='r', label='Testing Accuracy, Adam', legend=False, ax=ax2)
ax.set_xlabel("Epoch")
ax2.set_ylabel('Accuracy')
ax.set_ylabel('BCE Loss')
f.legend(loc='center')
plt.savefig(path+"/new1.png", dpi=600)
```
